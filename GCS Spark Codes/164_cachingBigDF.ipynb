```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('caching big DF').getOrCreate()
```

```python
df = spark.read.format('csv').option('header', True).load('/data/customers500.csv')
```

```python
df.first()
```

```python
df.show()
```

```python
df.count()
```

```python
df.cache()
```

```python
df.show()
```

```python
df2 = spark.read.format('csv').option('header', True).option('inferSchema', True).load('/data/customers10.csv')
```

```python
df2.cache() #it's lazy, so here no cache will create
```

```python
df2.show() #here action is there, so spark will finally execute the cache --> it will a bit slow as caching is happening here.. next df.show() will be a lot faaaaaster ✅
```

```python
df2.tail(10)
```

```python
df.unpersist()
df2.unpersist()
```

```python
spark.stop()
```
