```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Spark Creating DF').getOrCreate()
```

```python
data = [
    ('Alice', 'Mumbai', 35),
    ('Bob', 'Pune', 23),
    ('Charlie', 'Delhi', 43)
]
column = ['name', 'city', 'age']
```

```python
df = spark.createDataFrame(data, column)
```

```python
df.show()
```

### Spark SQL

```python
spark.sql('show tables').show()
```

```python
spark.sql("select * from external_customers where _c2 = 'Delhi'").show()
```

### Spark Table

```python
df_table = spark.table('external_customers')
df_table.show()
```

### spark range

```python
spark.range(0, 10).show() 
```

### RDD -> DF

```python
rdd = spark.sparkContext.parallelize([('Alice', 20), ('Bob', 23)])
rdd.first()
```

```python
df_rdd = rdd.toDF(['name', 'age'])
```

```python
df_rdd.show()
```
