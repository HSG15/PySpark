```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('SparkTable').getOrCreate()
```

```python
df = spark.read.format('csv').option('header', True).option('inferSchema', True).load('/data/customers1.csv')
```

```python
df.show(5)
```

```python
df.printSchema()
```

```python
spark.sql('show tables').show()
```

### Temporary Table -> createOrReplaceTempView()

```python
df.createOrReplaceTempView('temp_customers')
```

```python
spark.sql('show tables').show()
```

```python
spark.sql('select * from temp_customers limit 5').show()
```

### Global Temporary Table -> createOrReplaceGlobalTempView()

```python
df.createOrReplaceGlobalTempView('global_temp_customers')
```

```python
spark.sql('show tables in global_temp').show()
```

```python
spark.sql('drop table global_temp.temp_customers')
```

```python
spark.sql('show tables in global_temp').show()
```

```python
spark.sql('select * from global_temp.global_temp_customers limit 4').show(5)
```

### Persistent Table -> saveAsTable()

```python
df.write.mode('overwrite').saveAsTable('customers_persistent')
```

```python
spark.sql('show tables').show()
```

```python
# Access the SparkContext and then get the application ID
app_id = spark.sparkContext.applicationId

# Print the app ID
print(f"The application ID is: {app_id}")
```
