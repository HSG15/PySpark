```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('bigCaching').getOrCreate()
```

```python
hdfs_path = '/data/customers500.csv'
rdd = spark.sparkContext.textFile(hdfs_path)
```

```python
header = rdd.first()
```

```python
rdd2 = rdd.filter(lambda row : row != header).map(lambda row : row.split(','))
```

```python
rdd2.first()
```

```python
reducedCityRDD = rdd2.map(lambda row : (row[2], 1)).reduceByKey(lambda x, y : x+y)
```

```python
reducedCityRDD.collect()
```

### cache()

```python
reducedCityRDD.cache()
```

```python
reducedCityRDD.count()
```

```python
reducedCityRDD.persist().is_cached
```

```python
reducedCityRDD.unpersist() #To uncache
```

```python
spark.stop()
```
