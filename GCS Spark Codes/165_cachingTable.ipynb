```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('caching table').getOrCreate()
```

```python
path = '/data/customers10.csv'
df = spark.read.format('csv').option('header', True).load(path)
```

```python
df.head()
```

```python
spark.sql('show tables').show()
```

```python
df.createOrReplaceTempView('cache_table')
```

```python
spark.sql('select * from cache_table limit 5').show()
```

```python
spark.sql('show tables').show()
```

### Cache spark table

```python
spark.sql('show databases').show()
```

```python
spark.sql('SHOW TABLES FROM default').show()
```

```python
spark.sql('cache table cache_table') # Eager Caching
```

```python
spark.sql('select c.karnataka, c.india from cache_table c order by india').show()
```

```python
spark.sql('cache lazy table cache_table') # make the cache lazy
```

```python
spark.sql('select c.karnataka, c.india from cache_table c order by india').show()
```

```python
spark.sql("UNCACHE TABLE cache_table")
```
